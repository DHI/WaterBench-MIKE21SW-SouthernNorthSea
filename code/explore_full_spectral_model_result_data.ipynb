# Explore full spectral model results

Load the data using [MIKE IO](https://dhi.github.io/mikeio) and review it. In this notebook we consider the directional-frequency wave spectra at selected grid points. 

### Load packages

```python
import mikeio
import matplotlib.pyplot as plt 
plt.rcParams.update({'font.size': 10})
```

### Define paths

```python
#mr_fldr = "../output_sample/"
mr_fldr = "../output/" # Change the output folder if needed  
fig_fldr = "../figures/"
save_fig = True
show_fig = True
```

### Open the full spectral model result file

```python
fn = "Points_spectral.dfsu"
ds = mikeio.read(f"{mr_fldr}{fn}")
ds
```

### Plot energy density

```python
# Select time step and station and plot the spectral energy density
plot_time = "2022-05-05" if "2022-05-01" in ds.time else ds.time[-1]
station = "Europlatform2: Energy density"

fig = plt.figure(figsize=(5,5))
ds[station].sel(time=plot_time).plot(title=station)
if save_fig:
    plt.savefig(f"{fig_fldr}spectral_energy_density.png")
if show_fig:
    plt.show()
```
